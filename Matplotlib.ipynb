````python
chess = pd.read_csv('indian_grandmasters_july_2020.csv')
chess.head()
#plotting pie chart shows GM Propotion as state Wise
states_with_counts= chess['State'].value_counts() # obtaining states and number of GMs
state = states_with_counts.index.tolist() # seprating state for labeling
counts = states_with_counts.values.tolist() # poltting values
explode = [0.1 if label == 'Tamil Nadu' else 0 for label in state]
plt.style.use('_classic_test_patch') #using pre-defined styles just like using bootstraps and tailwind for css
plt.pie(counts,labels=state,shadow=True,explode = explode) #plotting pie chart
plt.show()
````

🧠 "Chess Leadership in India: Tamil Nadu’s Dominance"

This pie chart visualizes the state-wise distribution of Indian Chess Grandmasters as of July 2020. Tamil Nadu stands out as the epicenter of chess excellence, contributing the largest share of Grandmasters, followed by West Bengal and Maharashtra.
**The distribution showcases regional disparities, hinting at differences in access to coaching, infrastructure, and chess culture.**
This insight not only highlights where India’s chess talent flourishes but also where potential may lie untapped.

**Note:**  
While the pie chart gives a quick overview of each state’s share, small slices become difficult to compare accurately. In situations where you need to analyze low‑value categories in detail, switching to a bar chart is often much more effective.


````python
#bar chart for same scenerio
plt.style.use('ggplot')
plt.barh(state,counts,color='#77DD77')
plt.title('Chess Leadership in India :High to Low ')
plt.xlabel('Number of Grandmasters')
plt.show()
#we have more labels at the time we use bar chart it will overlap on one another so Horizontal bar chart is used
#we are gonna reverse list state and count so that we can get highest values from top
state.reverse()
counts.reverse()
plt.barh(state,counts,color='#87CEEB')
plt.title('Chess Leadership in India : Low to High')
plt.xlabel('Number of Grandmasters')
plt.tight_layout()#to avoid overlap of labels
plt.show()
````

# 🥧 Matplotlib Module 5 – Pie Charts, Histograms & Boxplots

###🥧 Pie Chart – Definition
A pie chart is used to show the proportion of categories in a dataset, where each slice represents a percentage of the total.
It is useful for visualizing part-to-whole relationships like expenses, population shares, or survey results.

####Key Features:

labels – category names

sizes – percentage values

explode – to highlight a slice

autopct – show % text

startangle – rotate starting slice

plt.axis('equal') – keeps the circle circular

###📊 Histogram – Definition
A histogram is used to represent the distribution of continuous data.
It divides data into equal-width intervals (bins) and counts how many values fall into each bin.

####Key Features:

Shows frequency of values

Detects skew, spread, and outliers

bins = number of intervals

Useful for test scores, heights, salaries, etc.

### 📦 Boxplot – Definition
A box plot (box-and-whisker plot) shows the spread and symmetry of data.
It is ideal for comparing multiple groups and spotting outliers.

###Key Features:

Box = 25th to 75th percentile (IQR)

Line inside = median

Whiskers = range of values

Dots = outliers

Use patch_artist=True to color the boxes

💡 Tip for All:
Add titles, grid lines, and axis labels to make your plots easier to read and interpret.




````python
#we already done pie chart once before its time to recall that concept once again
#this time we gonna analysis grandmaster per decade
years = (chess['Year_of_becoming_GM'])
decade = (years // 10)*10
counts = (decade.value_counts()).sort_values()
decade_list = counts.index.tolist()
counts_list = counts.values.tolist()
fig_5 = plt.figure(figsize=(8,4))
axis_5 = fig_5.add_axes([0.1,0.1,0.9,0.9])
explode = [ 0.1 if x== 2010 else 0 for x in decade_list]
axis_5.pie(counts_list,explode=explode,labels=decade_list,autopct='%1.1f%%',shadow=True)
axis_5.axis('equal')
plt.show()
````

"The Golden Decade of Indian Chess"
This pie chart clearly illustrates a phenomenal rise in Indian Grandmasters during the 2010s — a testament to improved infrastructure, digital learning, and national-level chess initiatives.
c68.2% of GMs in India earned their title in this single decade — a true golden era for Indian chess!



````python
#Histograms
#now we gonna analysis rateing distribution

#data processing
rateing = chess['Classical_Rating'].dropna()
rateing_list = rateing.values.tolist()

#plotting histogram
fig_hist = plt.figure(figsize=(8,4))
axis_hist = fig_hist.add_axes([0.1,0.1,0.9,0.9])
axis_hist.set_facecolor('gray')
axis_hist.hist(rateing_list,color='orange',)
axis_hist.set_title('Rateing Distribution')
axis_hist.set_xlabel('Rateing')
axis_hist.set_ylabel('Frequency')

# Add mean line
mean_rating = rateing.mean()
plt.axvline(mean_rating,color='blue', linestyle='--',linewidth=1,label=f'Mean = {mean_rating:.1f}')

plt.grid(color='pink')
plt.legend()
plt.show()
````

````python
#box_plots

plt.figure(figsize=(6, 5))

plt.boxplot(chess['Classical_Rating'].dropna(),boxprops={'color':'red'},
            medianprops={'color':'blue'},showfliers=True)
plt.title('Classical Rating Box Plot')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
````

🧱 What is a 3D Surface Plot?
A 3D surface plot represents a three-dimensional dataset, where:

X and Y form a grid (2D mesh of coordinates)

Z contains values evaluated at each (X, Y) point

The surface shows how Z varies over that 2D plane

| Parameter   | Description                              |
| ----------- | ---------------------------------------- |
| `X, Y, Z`   | 2D arrays (from `meshgrid`)              |
| `cmap`      | Colormap (e.g., `'viridis'`, `'plasma'`) |
| `edgecolor` | Edge line color (`'none'` for smooth)    |
| `alpha`     | Transparency (0 to 1)                    |
| `linewidth` | Grid line width                          |
